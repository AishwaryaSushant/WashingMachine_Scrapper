```python
import csv
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import requests
import PyPDF2
import io

# Function to extract data from pdf
def extracted_data_from_spec_sheet(spec_sheet_content):
    extracted_data = {}
    pdf_stream = io.BytesIO(spec_sheet_content)

    pdf = PyPDF2.PdfReader(pdf_stream)
    full_text = ""
    for page_num in range(len(pdf.pages)):
        page = pdf.pages[page_num]
        full_text += page.extract_text()
#     full_text = full_text.replace("\n", " ").replace("\r", " ").strip()

    return full_text

browser = webdriver.Chrome()
browser.maximize_window()
browser.get('https://www.lg.com/us/washers')

# Opening url for front load washers
model_name = 'front-load-washers' # use a variable to store the model name
browser.get(f'https://www.lg.com/us/{model_name}') # use f-string to insert the model name in the url

product_data = []
model = []
links = []

# Clicking on load more button until all products are loaded
while True:
    try:
        load_more = browser.find_element(By.XPATH, "//button[text()='Load More']")
        browser.execute_script("arguments[0].scrollIntoView();", load_more)
        browser.execute_script("arguments[0].click();", load_more)
        time.sleep(2) # Add a small delay to ensure products are loaded
    except:
        break

# Relocating elements each time after going back to previous page
prod_links = browser.find_elements(By.CLASS_NAME, "css-11xg6yi")
links_list = [link.get_attribute('href') for link in prod_links]

# Use a for loop to iterate over the urls and extract the data from each one
for link in links_list:
    browser.execute_script("window.open('" + link +"');")

    try:
        # Switching to the new window that opens after clicking the link
        new_window = browser.window_handles[-1]
        browser.switch_to.window(new_window)

        # Copying the model number 
        model_no = browser.find_element(By.XPATH,'//*[@id="__next"]/div[7]/div[1]/div[2]/div[1]/div[1]/span').text

        # Clicking on ALL SPECS if available
        WebDriverWait(browser, 30).until(EC.presence_of_element_located((By.XPATH, '//*[@id="__next"]/div[7]/div[1]/div[2]/div[1]/div[6]/div/div/div/div[3]/div/a')))
        
        all_specs = browser.find_element(By.XPATH, '//*[@id="__next"]/div[7]/div[1]/div[2]/div[1]/div[6]/div/div/div/div[3]/div/a')
        
        if all_specs.is_displayed():
            browser.execute_script("arguments[0].scrollIntoView();", all_specs)
            browser.execute_script("arguments[0].click();", all_specs)
            
            # Check if spec sheet is available
            spec_sheet = browser.find_element(By.XPATH, '//a[contains(text(),"DOWNLOAD SPEC SHEET")]')
            
            if spec_sheet.is_displayed():
                # Clicking on Download spec sheet
                spec_sheet.click()
                spec_sheet_url = spec_sheet.get_attribute("href")
                print(spec_sheet_url)
                
                if spec_sheet_url:
                    # Use a try-except block to handle any errors that might occur while downloading or parsing the pdf files
                    try:
                        spec_sheet_response = requests.get(spec_sheet_url)
                        spec_sheet_data = extracted_data_from_spec_sheet(spec_sheet_response.content)
                        print(model_no)
                        model.append(model_no)
                        print(spec_sheet_data)
                        print("\n")
                        product_data.append(spec_sheet_data) # append the extracted data to the list
                    except Exception as e:
                        print(f"Failed to process product, Error: {e}")
    except Exception as e:
        browser.back()
    finally:
        # Closing the new window and switching back to the main window
        browser.close()
        main_window = browser.window_handles[-1]
        browser.switch_to.window(main_window)
        # Using execute_script() with window.history.go(-1) instead of back()
        browser.execute_script("window.history.go(-1)")
        time.sleep(2) # Add a small delay before relocating elements

# Use pandas to create a dataframe and save it as csv or excel file
df = pd.DataFrame({'Model': model, 'Product Data': product_data})
csv_filename = 'front_load_specifications.csv'
df.to_csv(csv_filename, index=False)

print("\nProcess Complete")
```
